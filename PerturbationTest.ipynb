```python
import matplotlib.pyplot as plt
%matplotlib inline

import pandas
from GroupDecisionModel import Actor, Model
```

```python
df = pandas.read_csv("ExampleActors.csv")
df.Salience /= 100
```

## Basic Run

```python
model = Model.from_dataframe(df, {"name": "Actor",
                                  "x": "Position",
                                  "c": "Capability",
                                  "s": "Salience"}, verbose=False)
```

```python
model.run_model(10)
```

```python
original_medians = model.median_positions
```

```python
plt.plot(model.median_positions)
```

## Perturbed runs

```python
all_median_positions = []
for i in range(100):
    model = Model.from_dataframe(df, {"name": "Actor",
                                  "x": "Position",
                                  "c": "Capability",
                                  "s": "Salience"}, verbose=False)
    model.perturb_salience(0.05)
    model.run_model(10)
    all_median_positions.append(model.median_positions)
    
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111)
for series in all_median_positions:
    ax.plot(series, c='k', alpha=0.1)

ax.set_ylim(0, 11)
```

```python
step_distributions = []
for i in range(11):
    vals = []
    for series in all_median_positions:
        vals.append(series[i])
    step_distributions.append(vals)
    
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111)
boxes = ax.boxplot(step_distributions)

f = ax.plot([7] + original_medians, c='k')

ax.set_ylim(0, 11)
ax.set_xlabel("Step")
ax.set_ylabel("Median Position")
```
